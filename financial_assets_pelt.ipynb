```python
import yfinance as yf
import matplotlib.pyplot as plt
import ruptures as rpt
```

```python
def download_data(ticker, start_date, end_date):

    data = yf.Ticker(ticker).history(start=start_date, end=end_date, interval="1d")

    return data
```

```python
def detect_change_points(series, model="rbf", pen=3, min_size=3, jump=5):
    algo = rpt.Pelt(model=model, min_size=min_size, jump=jump).fit(series)
    breakpoints = algo.predict(pen=pen)
    return breakpoints
```

```python
def plot_change_points(series, breakpoints):
    rpt.display(series, breakpoints, figsize=(10, 6))
    plt.show()
```

```python
#Exemplo com o Bitcoin (BTC-USD)
ticker = "BTC-USD" #Código do Bitcoin em relação ao dolar
start_date = "2023-01-01"
end_date = "2024-12-31"

data = download_data(ticker, start_date, end_date)  # Baixar os dados

# Detectar pontos de mudança
close_prices = data["Close"].values
breakpoints = detect_change_points(close_prices)
print("Pontos de mudança detectados:", breakpoints)

# Visualizar os resultados
plot_change_points(close_prices, breakpoints)
```
